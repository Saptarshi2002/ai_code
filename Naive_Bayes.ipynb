```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('weather_data.csv')
df
```

```python
inputs = df.drop('PlayTennis', axis='columns')
target = df['PlayTennis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2, random_state=42)
X_train
# X_test
```

```python
from sklearn.preprocessing import LabelEncoder
le_outlook = LabelEncoder()
le_temp = LabelEncoder()
le_humidity = LabelEncoder()
le_wind = LabelEncoder()

inputs['Outlook_n'] = le_outlook.fit_transform(inputs['Outlook'])
inputs['Temp_n'] = le_temp.fit_transform(inputs['Temperature'])
inputs['Humidity_n'] = le_humidity.fit_transform(inputs['Humidity'])
inputs['Wind_n'] = le_wind.fit_transform(inputs['Wind'])

inputs
```

```python
inputs_n = inputs.drop(['Outlook', 'Temperature', 'Humidity', 'Wind'], axis='columns')
inputs_n
```

```python
model = GaussianNB()
model.fit(inputs_n,target)
```

```python
manual_input = [[2, 0, 1, 1]]
model.predict(manual_input)
```

```python
# 1. Define the specific weather condition
new_day_weather = pd.DataFrame({
    'Outlook': ['Sunny'],
    'Temperature': ['Hot'],
    'Humidity': ['High'],
    'Wind': ['Strong']
})

# 2. Transform the values FROM the dataframe
input_numbers = [
    le_outlook.transform(new_day_weather['Outlook'])[0],      # Reads 'Sunny'
    le_temp.transform(new_day_weather['Temperature'])[0],     # Reads 'Hot'
    le_humidity.transform(new_day_weather['Humidity'])[0],    # Reads 'High'
    le_wind.transform(new_day_weather['Wind'])[0]            # Reads 'Strong'
]

# 3. Predict
prediction = model.predict([input_numbers])
# predicted_class = le_target.inverse_transform(prediction)

# print(f"Prediction for this specific weather: {predicted_class[0]}")

prediction
```
